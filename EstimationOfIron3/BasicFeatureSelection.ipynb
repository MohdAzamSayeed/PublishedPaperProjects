## Feature Selection using Linear Regression

```python
dataset.info()
```

```python
dataset2=dataset
dataset2
```

```python
print(dataset2[dataset2['PaymentMethod'].isna()==True].size)
print(dataset2[dataset2['Contract'].isna()==True].size)
print(dataset2[dataset2['MonthlyCharges'].isna()==True].size)
```

```python
dataset2['Churn'].replace(['Yes','No'],[1,0],inplace=True)
dataset2['PaymentMethod'].replace(['Electronic check','Mailed check','Bank transfer (automatic)','Credit card (automatic)'],[0,1,2,3],inplace=True)
dataset2['Contract'].replace(['Month-to-month','One year','Two year'],[0,1,2],inplace=True)
```

```python
print(dataset2['Churn'].unique())
print(dataset2['PaymentMethod'].unique())
print(dataset2['Contract'].unique())
print(dataset2['MonthlyCharges'].unique())
```

```python
churn =pd.DataFrame(dataset2['Churn'])
PaymentMethod =pd.DataFrame(dataset2['PaymentMethod'])
Contract = pd.DataFrame(dataset2['Contract'])
MonthlyCharges = pd.DataFrame(dataset2['MonthlyCharges'])
```

```python
#Linear regression and find Accuracy of Prediction
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
import numpy as np
for xx in [PaymentMethod,Contract,MonthlyCharges] :
    X_train,X_test, Y_train,Y_test =train_test_split(xx,churn,test_size=0.2,random_state=1)
    regressor = LogisticRegression()
    regressor.fit(X_train,Y_train)
    y_pred = regressor.predict(X_test)
    y_pred = pd.DataFrame(y_pred,columns=['Predicted'])
    print('Accuracy: ', (regressor.score(y_pred, Y_test)))
    print('Root Mean Square Error: ',np.sqrt(metrics.mean_squared_error(Y_test,y_pred)) )
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```
