```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv('BigMartSalesData.csv')
```

```python
data.head()
```

1.Plot Total Sales Per Month for Year 2011.  How the total sales haveincreased over months in Year 2011. Which month has lowest Sales?

```python
d2011=data[data['Year']==2011]
```

```python
dline=d2011.groupby('Month').sum().reset_index()
```

```python
x=dline['Month']
y=dline['Amount']
```

```python
plt.plot(x,y)
plt.show()
#lowest sales in Month of feb in year 2011
```

2.Plot Total Sales Per Month for Year 2011 as Bar Chart.  Is Bar Chart Better to visualize than Simple Plot?est towards sales?


```python
plt.bar(x,y)
```

3.Plot Pie Chart for Year 2011 Country Wise. Which Country contributes high

```python
dPie=d2011.groupby('Country').sum().reset_index()
dPie
```

```python
plt.rcParams['figure.figsize'] =(10,10)
plt.pie(dPie['Quantity'], labels=dPie['Country'], autopct='%1.1f%%',
        shadow=True, startangle=90)
#plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

4.Plot Scatter Plot for the invoice amounts and see the concentration of amount.  In which range most of the invoice amounts are concentrated

```python
plt.rcParams['figure.figsize'] =(4,5)
plt.scatter(x,y)
plt.xlabel('month')
plt.ylabel('amount')
plt.show()
```
