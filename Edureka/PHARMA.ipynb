Maple Leaves Ltd is a start-up company which makes herbs from different types of plants and its leaves. Currently the system they use to classify the trees which they import in a batch is quite manual. A labourer from his experience decides the leaf type and subtype of plant family. They have asked us to automate this process and remove any manual intervention from this process.Objective: To classify the plant leaves by various classifiers from different metrics of the leaves and to choose the best classifier for future reference.

2)Import the required classification libraries along with pandas, numpy, seaborn etc

```python
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_validate
```

3)Then import the classifiers from them (Randomforest, SVM, NaiveBayes, DecisionTrees)

```python
randomfC=RandomForestClassifier()
svnC=svm.SVC(kernel='linear',probability=True)
nbC=GaussianNB()
dtC=DecisionTreeClassifier()
```

4)After this create a function to encode the labels of the strings given in the dataset

```python
traindata['species'].unique()
```

```python
global le
def Encoder(df,col):
    le = LabelEncoder()
    le.fit(df[str(col)].unique())
    traindata[str(col)]=le.transform(traindata[str(col)])
```

```python
Encoder(traindata,'species')
```

5)You can do the above step using label encoder. With this you are creating some labels from train set as test set. The test set we imported is for testing the best classifier accuracy once we choose it

6)Then extract the values from train set by stratifying them and dividing it into 80:20 ratio

7)Now your X train, X test, Y train, Y test are ready.

```python
train,test=train_test_split(traindata,test_size=0.2,random_state=4)
print(train.shape)
print(test.shape)
```

```python
x=traindata.iloc[:,1:]
y=traindata.iloc[:,0]
```

```python
train_x=train.iloc[:,1:]
train_y=train.iloc[:,0]
test_x=test.iloc[:,1:]
test_y=test.iloc[:,0]
```

8)We currently donâ€™t know which is the best classifier on the dataset. So,we apply all 4 of them.

9)Create the classifiers class and initialize all the respective classifiers

10)Then run the X train & X test datasets through classifiers calculating the log loss and accuracy of the result

```python
randomfC.fit(train_x,train_y)
pred=randomfC.predict(test_x)
print('Accuracy:',metrics.accuracy_score(pred,test_y))
yk_grd_probs = randomfC.predict_proba(test_x)
print('Log Loss:',metrics.log_loss(test_y, yk_grd_probs, labels=traindata['species'].unique()))
print('Cross validate:',cross_validate(randomfC,x,y,cv=10)['test_score'].mean())
```

```python
svnC.fit(train_x,train_y)
pred=svnC.predict(test_x)
print('Accuracy:',metrics.accuracy_score(pred,test_y))
yk_grd_probs = svnC.predict_proba(test_x)
print('Log Loss:',metrics.log_loss(test_y, yk_grd_probs, labels=traindata['species'].unique()))
print('Cross validate:',cross_validate(svnC,x,y,cv=10)['test_score'].mean())
```

```python
nbC.fit(train_x,train_y)
pred=nbC.predict(test_x)
print('Accuracy:',metrics.accuracy_score(pred,test_y))
yk_grd_probs = nbC.predict_proba(test_x)
print('Log Loss:',metrics.log_loss(test_y, yk_grd_probs, labels=traindata['species'].unique()))
print('Cross validate:',cross_validate(nbC,x,y,cv=10)['test_score'].mean())
```

```python
dtC.fit(train_x,train_y)
pred=dtC.predict(test_x)
print('Accuracy:',metrics.accuracy_score(pred,test_y))
yk_grd_probs = dtC.predict_proba(test_x)
print('Log Loss:',metrics.log_loss(test_y, yk_grd_probs, labels=traindata['species'].unique()))
print('Cross validate:',cross_validate(dtC,x,y,cv=10)['test_score'].mean())
```
