```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df = pd.read_csv('BX-Book-Ratings.csv',encoding='latin1')
df=df[df['rating']==10]  #higher rentals
df.head()
```

```python
len(df)
```

```python
df['isbn'] = df['isbn'].str.strip()
df.drop(columns=['rating'], inplace=True)
#df['user_id'] = df['user_id'].astype('str')
#df['isbn'] = df['isbn'].astype('str')
```

```python
df=df.head(10) #filtered based on highest rating and the selecting top 10 only due to system performance please include full 
#data set if you have good system performance
df
```

```python
basket = (df.groupby(['user_id', 'isbn'])['isbn']
          .count().unstack().reset_index().fillna(0)
          .set_index('user_id'))
basket
```

```python
frequent_itemsets = apriori(basket, min_support=0.07, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.head()
```

```python
rules[ (rules['lift'] >= 6) &
      (rules['confidence'] >= 0.8) ]
```
