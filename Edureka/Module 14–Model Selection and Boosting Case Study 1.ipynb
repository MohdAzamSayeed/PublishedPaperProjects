```python
import pandas as pd
```

```python
data=pd.read_csv('bio-degradabale-data.csv',sep=';',header=None)
data.rename(columns={41: "Output"},inplace=True)
data.head()
```

```python
data.isna().sum()
```

```python
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn import metrics
train,test=train_test_split(data,test_size=0.3,random_state=4)
print(train.shape)
print(test.shape)
```

```python
train_x=train.iloc[:,:-1]
train_y=train.iloc[:,-1]
test_x=test.iloc[:,:-1]
test_y=test.iloc[:,-1]
```

```python
from sklearn.ensemble import AdaBoostClassifier
model=AdaBoostClassifier(n_estimators=30,random_state=7)
model.fit(train_x,train_y)
y_pred=model.predict(test_x)
metrics.accuracy_score(test_y,y_pred)
```

```python
data2=pd.read_csv('bio-degradabale-data.csv',sep=';',header=None)
data2.rename(columns={41: "Output"},inplace=True)
data2.head()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(data2['Output'])
data2['Output']=le.transform(data2['Output'])
```

```python
data2.head()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(data2.iloc[:,:-1])
existing_2d = pca.transform(data2.iloc[:,:-1])
existing_df_2d = pd.DataFrame(existing_2d)
existing_df_2d.index = data2.index
existing_df_2d.columns = ['PC1','PC2']
existing_df_2d.head()
```

```python
existing_df_2d['Ouput']=data2['Output']
```

```python
train,test=train_test_split(existing_df_2d,test_size=0.3,random_state=4)
print(train.shape)
print(test.shape)
train_x=train.iloc[:,:-1]
train_y=train.iloc[:,-1]
test_x=test.iloc[:,:-1]
test_y=test.iloc[:,-1]
```

```python
model=AdaBoostClassifier(n_estimators=30,random_state=7)
model.fit(train_x,train_y)
y_pred=model.predict(test_x)
metrics.accuracy_score(test_y,y_pred)
```
