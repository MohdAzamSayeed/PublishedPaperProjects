```python
import pandas as pd
data = pd.read_csv('sample.csv')
```

```python
data.head()
```

```python
X= data.drop(['target'], axis=1)
Y=data['target']
```

```python
X.columns[:]
```

```python
X1 =data[['R','G','B','y3','cb','cr']]
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(X1)
scaled_data = pd.DataFrame(scaled_data, columns=X1.columns)
scaled_data.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(X)
scaled_data = pd.DataFrame(scaled_data, columns=X.columns)
scaled_data.head()
```

```python
X['R'] =scaled_data['R']
```

```python
X['G']= scaled_data['G']
X['B']= scaled_data['B']
X['y3']= scaled_data['y3']
X['cb']= scaled_data['cb']
X['cr']= scaled_data['cr']
X.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.3)
```

```python
#pip install xgboost
import xgboost as xgb
xgbc=xgb.XGBClassifier(n_estimator=10,max_dept=5)
```

```python
xgbc.fit(x_train,y_train)
```

```python
y_pred=xgbc.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
xgbc.
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
feature_imp = pd.Series(xgbc.feature_importances_,X.columns).sort_values(ascending=False)
feature_imp
```

```python
import matplotlib.pyplot as plt
import numpy as np

features = X.columns
importances = xgbc.feature_importances_
indices = np.argsort(importances)
plt.figure(figsize=(10,5))

plt.title('Feature Importances')
plt.barh(range(len(indices)), importances[indices], color='b', align='center')
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.xlabel('Relative Importance')
plt.show()
```
